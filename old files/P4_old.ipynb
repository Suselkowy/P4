```python
import pandas as pd

def process_csv(file_path):
    # Load the CSV file
    df = pd.read_csv(file_path)
    
    # Strip whitespace from column names
    df.columns = df.columns.str.strip()
    
    # Show unique values of the 'Label' column
    unique_labels = df['Label'].unique()
    
    # Count rows with BENIGN label
    benign_count = (df['Label'] == 'BENIGN').sum()
    
    # Count rows with label different than BENIGN
    non_benign_count = (df['Label'] != 'BENIGN').sum()
    
    return df, unique_labels, benign_count, non_benign_count

# Example usage
# Replace 'your_file.csv' with the actual path to your CSV file
df, unique_values, benign_rows, non_benign_rows = process_csv('Wednesday-workingHours.pcap_ISCX.csv')

print("Unique values in Label column:")
print(unique_values)
print(f"\nNumber of rows with BENIGN label: {benign_rows}")
print(f"Number of rows with non-BENIGN label: {non_benign_rows}")
print(f"Total rows: {len(df)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def process_csv_and_plot(file_path):
    # Load the CSV file
    df = pd.read_csv(file_path, encoding='cp1252')
    
    # Strip whitespace from column names
    df.columns = df.columns.str.strip()
    
    # Get value counts for all labels
    label_counts = df['Label'].value_counts()
    
    # Create the plot
    plt.figure(figsize=(12, 8))
    bars = plt.bar(label_counts.index, label_counts.values, color='skyblue', edgecolor='black')
    
    # Customize the plot
    plt.title('Count of Each Label', fontsize=16, fontweight='bold')
    plt.xlabel('Label', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    
    # Add value labels on top of bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{int(height)}',
                ha='center', va='bottom')
    
    # Adjust layout to prevent label cutoff
    plt.tight_layout()
    plt.show()
    
    # Print the counts
    print("Label counts:")
    for label, count in label_counts.items():
        print(f"{label}: {count}")
    
    return label_counts
```

```python
label_counts = process_csv_and_plot('Wednesday-workingHours.pcap_ISCX.csv')
```

```python
import pandas as pd
import os

# Directory containing the CSV files
directory = 'packet_fields_csv'

# List to hold filtered dataframes
filtered_dfs = []

# Loop through files Packet_Fields_File_X.csv where X is 1 to 18
for i in range(1, 19):
    file_name = f'Packet_Fields_File_{i}.csv'
    file_path = os.path.join(directory, file_name)
    
    if os.path.exists(file_path):
        try:
            # Read individual file
            df = pd.read_csv(file_path)
            print(f"Reading {file_name} with {len(df)} rows")
            
            # Filter to one random packet per flow_id for this file
            filtered_df = df.groupby('flow_id').apply(lambda x: x.sample(1)).reset_index(drop=True)
            
            # Add to list
            filtered_dfs.append(filtered_df)
            
            print(f"  -> Filtered to {len(filtered_df)} unique flow_ids")
            
        except Exception as e:
            print(f"Error processing {file_path}: {e}")
    else:
        print(f"File not found: {file_name}")

# Combine all filtered dataframes
if filtered_dfs:
    final_df = pd.concat(filtered_dfs, ignore_index=True)
    
    # Save the final dataset
    output_file = 'filtered_packet_fields.csv'
    final_df.to_csv(output_file, index=False)
    
    print(f"\nFinal dataset saved as '{output_file}'")
    print(f"Total rows in final dataset: {len(final_df)}")
    print(f"Files processed: {len(filtered_dfs)}")
    print(f"Columns: {list(final_df.columns)}")
    
else:
    print("No files were successfully processed.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the filtered dataset
df = pd.read_csv('filtered_packet_fields.csv')

print(f"Dataset loaded with {len(df)} rows")
print(f"Columns: {list(df.columns)}")

# Check the attack_label column
print(f"\nLabel distribution:")
print(df['attack_label'].value_counts())

# Set up the plotting style
plt.style.use('default')
sns.set_palette("husl")

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Attack Label Analysis', fontsize=16, fontweight='bold')

# 1. Bar plot of label counts
ax1 = axes[0, 0]
label_counts = df['attack_label'].value_counts()
bars = ax1.bar(range(len(label_counts)), label_counts.values)
ax1.set_xlabel('Attack Labels')
ax1.set_ylabel('Count')
ax1.set_title('Attack Label Distribution (Bar Chart)')
ax1.set_xticks(range(len(label_counts)))
ax1.set_xticklabels(label_counts.index, rotation=45, ha='right')

# Add value labels on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom')

# 2. Pie chart of label distribution
ax2 = axes[0, 1]
colors = plt.cm.Set3(range(len(label_counts)))
wedges, texts, autotexts = ax2.pie(label_counts.values, labels=label_counts.index, 
                                   autopct='%1.1f%%', colors=colors, startangle=90)
ax2.set_title('Attack Label Distribution (Pie Chart)')

# 3. Horizontal bar chart
ax3 = axes[1, 0]
y_pos = range(len(label_counts))
bars_h = ax3.barh(y_pos, label_counts.values)
ax3.set_yticks(y_pos)
ax3.set_yticklabels(label_counts.index)
ax3.set_xlabel('Count')
ax3.set_title('Attack Label Distribution (Horizontal Bar)')

# Add value labels on horizontal bars
for i, bar in enumerate(bars_h):
    width = bar.get_width()
    ax3.text(width + 0.1, bar.get_y() + bar.get_height()/2.,
             f'{int(width)}', ha='left', va='center')

# 4. Count plot using seaborn
ax4 = axes[1, 1]
sns.countplot(data=df, y='attack_label', ax=ax4, order=label_counts.index)
ax4.set_title('Attack Label Distribution (Seaborn Count Plot)')
ax4.set_xlabel('Count')

plt.tight_layout()
plt.show()

# Print summary statistics
print(f"\nSummary Statistics:")
print(f"Total unique labels: {df['attack_label'].nunique()}")
print(f"Most common label: {label_counts.index[0]} ({label_counts.iloc[0]} occurrences)")
print(f"Least common label: {label_counts.index[-1]} ({label_counts.iloc[-1]} occurrences)")

# Create a simple single chart version if you prefer
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='attack_label', order=df['attack_label'].value_counts().index)
plt.title('Attack Label Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Attack Labels')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

# Add count labels on top of bars
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width()/2., p.get_height()), 
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
